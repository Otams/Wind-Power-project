```python
len(data)
```

This data is a time sequence data therefore the traditional sequence of train_test split will not work, because the training has to be done in sequence and not picking random smaples 

```python
X_train, X_test, y_train,y_test = data['WindSpeed'][0:78000], data['WindSpeed'][78000:], data['ActivePower'][0:78000],  data['ActivePower'][78000:]
len(X_train),len(X_test),len(y_train),len(y_test)
```

Choice of model: extreme gradient boosting.

Gradient boosting according to Nvidia comes from the idea of boosting or improving a single weak model by combining it with several other weak models. This collectively will allow us to generate a strong model.

It's basically a concept of ensemble training, using multiple machine learning models in parallel on the same training data to iteratively set targets for the next model to minimize the error.

So extreme gradient boosting basically deals with decision trees, and in this particular model, which is xgb, they are using multiple decision trees to minimize the error and get a very high accuracy.

Implementing XGBoost: after data processing, it can be observed  that the date time features which we will see later, Wind speed variable was highly correlated with our target variable. `Active power`. This forms the basis of our problem or regression task. To implement `XGBoost` we'll use the library XGBoost, if not already installed the command `pip install XGBoost` can help install it into the environment.

Wind speed being the only feature variable of interest, a model can be built to  train on our data set, and the model will predict the last few samples, which in this case is 989 samples by taking that as test data.

```python
model = xgb.XGBRegressor(n_estimators = 200, early_stopping_rounds=50)
```

```python
model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)],
          verbose=True) 
```

```python
pred = model.predict(X_test)
```

```python
df_final = pd.DataFrame(data={'Actuals':y_test, 'Predictions':pred })  
```

```python
from sklearn.metrics import *
print('The Coefficient of determination (R-squared) = {:.3f}'.format(r2_score(df_final['Actuals'],df_final['Predictions'])))
print('The mean absolute error (MAE)                = {:.2f}'.format(mean_absolute_error(df_final['Actuals'],df_final['Predictions'])))
print('The RMSE error (RMSE)                        = {:.2f}'.format(mean_squared_error(df_final['Actuals'],df_final['Predictions'],squared=False)))
print('The Mean absolute percentage error (MAPE)    = {:.3f}'.format(mean_absolute_percentage_error(df_final['Actuals'],df_final['Predictions'])))  
```

```python
plt.plot(df_final['Predictions'],label='Prediction')
plt.plot(df_final['Actuals'], color='red',label='Actual')

plt.ylabel('Active Power')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

# New test only dates as features

So now that we have created a model where we trained it on the wind speed feature variable.

What if we didn't have wind speed and we wanted to predict future unseen values. That is dates that are not included or are after the actual data set that we have. For those since we don't have, uh, the wind speed value, we'll have to generate our own features using the date time that is given in the data set, and train the model to predict the target variable, which is active power.

Now, this won't be as accurate as using wind speed since the wind speed and active power are highly correlated. But we can still get the trend.
Let us look at how we can do that and predict future unseen values.

So before anything else, we'll first create an empty data frame with future dates. The last date in our data is 30th of March 2020. We can create a new data frame with dates ranging from 1st of April 2020 to 1st of June 2020, with frequency equal to ten minute is basically telling the function date range to create values at every ten minutes. And then we are passing that to a data frame as index.


```python
data.tail()
```

```python
new = pd.date_range('2020-04-01 12:00:00+00:00','2020-06-01 12:00:00+00:00', freq='10min')
new = pd.DataFrame(index=new)
```

```python
len(new)
```

We need to generate new features for our model to train on. What features can we generate?
We only have the dates and the hour and the minute so we can create, and we can use that to extract new features. Basically we'll be extracting the hour, minute, day, month, year, day of week, day of year, and week of year and generate new feature values.

So this function below deals with that particular process.

Since the index is generated by the date range sub function of pandas, we can call the hour, the minute and the day, and so on to create new values.
And these are basically making new columns in the already existing new dataframe.


```python
def create_f(df):
    df = df.copy()
    df['hour'] = df.index.hour
    df['minute'] = df.index.minute
    df['day'] = df.index.day
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofweek'] = df.index.dayofweek
    df['dayofyear'] = df.index.dayofyear
    df['weekofyear'] = df.index.isocalendar().week
    
    return df
```

```python
data2 = data.copy()
data2 = create_f(data2)
```

```python
data2.head()
```

```python
data2.columns
```

```python
data2.isna().sum()
data2 = data2.dropna()
```

```python
features = ['hour', 'minute', 'day', 'month', 'year','dayofweek', 'dayofyear']
label = 'ActivePower'
X_train, X_test, y_train,y_test = data2[features][0:78000], data2[features][78000:], data2[label][0:78000],  data2[label][78000:]
len(X_train),len(X_test),len(y_train),len(y_test)
```

```python
model2 = xgb.XGBRegressor(n_estimators = 1000,learning_rate=0.01)
model2.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)],
          verbose=True) 
```

```python
pred = model2.predict(X_test)
```

```python
df_final2 = pd.DataFrame(data={'Actuals':y_test, 'Predictions':pred})  
```

```python
from sklearn.metrics import *
print('The Coefficient of determination (R-squared) = {:.3f}'.format(r2_score(df_final2['Actuals'],df_final2['Predictions'])))
print('The mean absolute error (MAE)                = {:.2f}'.format(mean_absolute_error(df_final2['Actuals'],df_final2['Predictions'])))
print('The RMSE error (RMSE)                        = {:.2f}'.format(mean_squared_error(df_final2['Actuals'],df_final2['Predictions'],squared=False)))
print('The Mean absolute percentage error (MAPE)    = {:.3f}'.format(mean_absolute_percentage_error(df_final2['Actuals'],df_final2['Predictions'])))  
```

```python
plt.figure(figsize=(15,2))
sns.lineplot(data = data[label][74000:])
sns.lineplot(data = df_final2['Predictions'])

plt.figure(figsize=(15,2))
sns.lineplot(data = data[label][74000:].resample('D').mean())
sns.lineplot(data = df_final2['Predictions'].resample('D').mean())
```

```python
new = create_f(new)
new[features].head()
```

```python
fut = model2.predict(new[features])
future = pd.DataFrame(data={'Predictions':fut}, index = new.index)  
```

```python
plt.figure(figsize=(15,2))
sns.lineplot(data = data[label][74000:])
sns.lineplot(data = future['Predictions'])

plt.figure(figsize=(15,2))
sns.lineplot(data = data[label][74000:].resample('D').mean())
sns.lineplot(data = future['Predictions'].resample('D').mean())
```
